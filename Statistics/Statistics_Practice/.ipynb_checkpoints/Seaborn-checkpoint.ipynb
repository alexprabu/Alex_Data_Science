# Bar Graph

```python
flight_filepath = r"C:\Users\Alex\Data_Science\Data_Science_Course\Statistics\data-for-datavis\flight_delays.csv"
```

```python
flight_df = pd.read_csv(flight_filepath, index_col = "Month")
```

```python
flight_df.head()
```

```python
flight_df.tail()
```

```python
flight_df.isna.sum()
```

```python
plt.figure(figsize=(10,6))
plt.ylabel("Arrival delay (in minutes)")
sns.barplot(x=flight_df.index, y = flight_df['OO'])
```

# Heatmap

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=flight_df, annot = True)
```

# Correlation Heat Map

```python
plt.figure(figsize=(14,7))
corr = flight_df.corr()
ax = sns.heatmap(corr, vmin = -1, vmax = 1, center = 0, annot = True)
```
